```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import xgboost
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
from sklearn.model_selection import train_test_split
```

```python
cancer_df['diagnosis']=cancer_df['diagnosis'].replace('M', 1)
cancer_df['diagnosis']= cancer_df['diagnosis'].replace('B', 0)
```

```python
cancer_df.head()
```

```python
training_features=['texture_mean', 'area_mean', 'smoothness_mean', 'concavity_mean',
       'area_se', 'concavity_se', 'fractal_dimension_se', 'smoothness_worst',
       'concavity_worst', 'symmetry_worst']
```

```python
cancer_df_new= cancer_df[training_features]
```

```python
X=cancer_df[training_features]
y=cancer_df.diagnosis
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
logistic_model = LogisticRegression()      
logistic_model = logistic_model.fit(x_train,y_train)
logistic_predict = logistic_model.predict(x_test)
logistic_accuracy = accuracy_score(y_test,logistic_predict)
print('Accuracy is: ',100* logistic_accuracy)
logistic_confusion_matrix = confusion_matrix(y_test,logistic_predict)
sns.heatmap(logistic_confusion_matrix,annot=True,fmt="d")
```

```python
#Evaluating the model
print(classification_report(y_test, logistic_predict))
```
